```python
 df = pd.read_parquet('flights.parquet')
 df
```

```python
filtered_df = df[(df['DISTANCE'] < 1600) & (df['SCHEDULED_DEPARTURE'] >= 15.0) & (df['SCHEDULED_DEPARTURE'] < 16.0)]
print(filtered_df.shape)

# Calculate the Pearson correlation between DEPARTURE_DELAY and ARRIVAL_DELAY
# correlation = filtered_df['DEPARTURE_DELAY'].corr(filtered_df['ARRIVAL_DELAY'])

# print(f"Pearson correlation between DEPARTURE_DELAY and ARRIVAL_DELAY: {correlation}")
```

```python
Q1 = filtered_df['ARRIVAL_DELAY'].quantile(0.25)
Q3 = filtered_df['ARRIVAL_DELAY'].quantile(0.75)

# 3. Calculate IQR
IQR = Q3 - Q1

# Calculate lower bound and upper bound
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# 4. Count the number of outliers
outliers = filtered_df[(filtered_df['ARRIVAL_DELAY'] < lower_bound) | (filtered_df['ARRIVAL_DELAY'] > upper_bound)]

# Get the number of outliers
num_outliers = outliers.shape[0]

print(f"Number of outliers in ARRIVAL_DELAY: {num_outliers}")
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

# Assuming your DataFrame is already loaded as df

# 1. Filter the DataFrame for 'Furnished' houses in Mumbai and Delhi
mumbai_furnished = df[(df['Furnishing Status'] == 'Semi-Furnished') & (df['City'] == 'Hyderabad')]
delhi_furnished = df[(df['Furnishing Status'] == 'Semi-Furnished') & (df['City'] == 'Bangalore')]

# 2. Linear regression for Mumbai
X_mumbai = mumbai_furnished[['Size']]  # Feature (Size)
y_mumbai = mumbai_furnished['Rent']  # Target (Rent)

# Initialize and fit the linear regression model for Mumbai
model_mumbai = LinearRegression()
model_mumbai.fit(X_mumbai, y_mumbai)

# Slope (Rs/sqft) for Mumbai
slope_mumbai = model_mumbai.coef_[0]

# 3. Linear regression for Delhi
X_delhi = delhi_furnished[['Size']]  # Feature (Size)
y_delhi = delhi_furnished['Rent']  # Target (Rent)

# Initialize and fit the linear regression model for Delhi
model_delhi = LinearRegression()
model_delhi.fit(X_delhi, y_delhi)

# Slope (Rs/sqft) for Delhi
slope_delhi = model_delhi.coef_[0]

# 4. Calculate the absolute difference between the slopes
slope_difference = abs(slope_mumbai - slope_delhi)

print(f"Rent increase per square foot for Furnished houses in Mumbai: {slope_mumbai} Rs/sqft")
print(f"Rent increase per square foot for Furnished houses in Delhi: {slope_delhi} Rs/sqft")
print(f"Absolute difference between slopes: {slope_difference} Rs/sqft")
```

```python
# Group data to calculate barbeque percentage per city
bbq_percentage = generalinfo_df.groupby('city').apply(
    lambda x: (x['food_type'] == 'cajun').sum() * 100 / len(x)
).reset_index(name='bbq_percentage')

# Find San Francisco's barbeque percentage
sf_bbq_percentage = bbq_percentage[bbq_percentage['city'] == 'calistoga']['bbq_percentage'].values[0]

# Find cities with higher barbeque percentage than San Francisco
cities_above_sf = bbq_percentage[bbq_percentage['bbq_percentage'] > sf_bbq_percentage]

# Print the result
print(f"Number of cities with higher BBQ percentage than San Francisco: {len(cities_above_sf)}")
```
